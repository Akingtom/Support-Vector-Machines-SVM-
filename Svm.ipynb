```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Social_Network_Ads.csv')
df.head()
```

```python
df.shape
```

```python
X = df.iloc[:,[2,3]]
y = df.iloc[:, 4]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

#Applying Support Vector machine


```python
#svm
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
Y_pred = classifier.predict(X_test)
Y_pred
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, Y_pred))
```

```python
##using rbf
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0)
classifier.fit(X_train, y_train)

Y_pred = classifier.predict(X_test)
Y_pred

from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, Y_pred))
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', gamma = 1, C=7, random_state = 0)
classifier.fit(X_train, y_train)

Y_pred = classifier.predict(X_test)
Y_pred

print("Accuracy:",metrics.accuracy_score(y_test, Y_pred))
Y_pred = classifier.predict(X_test)
```

```python
svc = SVC(kernel='poly', degree=4, random_state=42)
svc.fit(X_train,y_train)

y_pred = svc.predict(X_test)

print('Accuracy Score:with poly kernel and degree')
print(metrics.accuracy_score(y_test,y_pred))
y_pred = svc.predict(X_test)
```

```python
#plot data points

# Use y_test for coloring and remove the undefined C
plt.scatter(X_test[:,0], X_test[:,1], c=y_test) # Plot all test points, not just the second one

#create the hyperplane
# Use the classifier trained with the linear kernel for plotting the linear hyperplane

# Re-train a linear SVC classifier or ensure the 'classifier' variable holds the linear model
# Based on your earlier code, you did train a linear SVC and assigned it to 'classifier'
# However, subsequent cells overwrote 'classifier' with RBF and polynomial kernels.
# To ensure you plot the linear hyperplane, re-initialize and train a linear SVC:
from sklearn.svm import SVC
linear_classifier = SVC(kernel = 'linear', random_state = 0)
linear_classifier.fit(X_train, y_train)

w = linear_classifier.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-2.5, 2.5)
yy = a * xx - (linear_classifier.intercept_[0]) / w[1]

# plot the hyperplane

plt.plot(xx, yy)
plt.axis('off'),  plt.show();
```
